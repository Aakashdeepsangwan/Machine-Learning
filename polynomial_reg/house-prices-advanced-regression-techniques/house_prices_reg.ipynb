```python
columns = ['PoolQC','MiscFeature','Alley','Fence'] 
train_df.drop(columns = columns, inplace = True) 
```

```python
train_df
```

```python
null_percent = train_df.isnull().mean() * 100
null_percent[null_percent > 0].sort_values(ascending=False)
```

```python
train_df['MasVnrType'] = train_df['MasVnrType'].replace(np.nan,'Unknown') 
```

```python
train_df['FireplaceQu'] = train_df['FireplaceQu'].replace(np.nan,'Unknown') 
```

```python
train_df['LotFrontage'].unique()
```

```python
# fill LOTFRONTAGE with median grouped by neighborhood
# houses in same neighborhood likely have similar frontage
train_df['LotFrontage'] = train_df.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median()))
```
